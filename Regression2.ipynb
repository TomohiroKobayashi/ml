```python
import re
def change_zen_to_han(teams):
    zen = list('ＡＢＣＤＥＦＧＨＩＪＫＬＭＮＯＰＱＲＳＴＵＶＷＸＹＺ'+'ａｂｃｄｅｆｇｈｉｊｋｌｍｎｏｐｑｒｓｔｕｖｗｘｗｚ')
    han = list('ABCDEFGHIJKLMNOPQRSTUVWXYZ' + 'abcdefghijklmnopqrstuvwxyz')
    
    zen_to_han = {}
    
    pattern = "|".join(zen)
    
    for i in teams:
        if re.search(pattern, i) is None:
            continue
        target = i*1
        for z, h in zip(zen, han):
            target = target.replace(z,h)
        zen_to_han[i] = target
    return zen_to_han

zen_to_han = change_zen_to_han(teams)
all_df['home_team'].replace(zen_to_han, inplace=True)
all_df['away_team'].replace(zen_to_han, inplace=True)
```
